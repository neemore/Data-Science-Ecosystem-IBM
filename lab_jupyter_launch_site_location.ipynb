Why this matters:

The Equator is at 0° latitude

At the Equator, Earth rotates at ~1,670 km/h (1,037 mph)
This rotational velocity decreases as you move toward the poles

Benefits of being closer to the Equator:

Free velocity boost - Earth's rotation gives rockets extra eastward velocity
Fuel savings - Less fuel needed to reach orbital velocity
Payload capacity - Can launch heavier payloads with same rocket
Cost efficiency - Reduces overall mission costs

CCAFS and KSC sites are on Florida's eastern coast
VAFB is on California's western coast

Why coastal locations are critical:
Safety Reasons:

Flight path over water - Rockets launch over the ocean, not populated areas
Failed launch protection - If a rocket fails, debris falls into ocean
Range safety - Minimizes risk to civilians and property
Downrange tracking - Easier to track and monitor over water

Logistical Reasons:

Sea transportation - Large rocket stages and components shipped by barge
Unlimited downrange - Ocean provides unobstructed flight path
Recovery operations - Easier to recover boosters and fairings from water
No airspace conflicts - Less commercial air traffic over ocean

Operational Reasons:

Polar orbits (VAFB) - Launch south over Pacific for polar trajectories
Equatorial orbits (Florida) - Launch east over Atlantic for equatorial orbits
Environmental buffer - Ocean acts as natural safety zone

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

def assign_marker_color(launch_outcome):
    if launch_outcome == 1:
        return 'green'
    else:
        return 'red'
    
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)
spacex_df.tail(10)
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

for index, record in spacex_df.iterrows():
    coordinate = [record['Lat'], record['Long']]
    marker = folium.Marker(
        coordinate,
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=folium.Popup(f"{record['Launch Site']}<br>Class: {record['class']}", max_width=200)
    )
    
    marker_cluster.add_child(marker)

site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
